```python
# q2
A = np.array([[0,-1,1],
              [1,-2,1],
              [0,1,-1]])
B = np.array([[1,0],
              [1,1],
              [1,2]])
C = np.array([1,1,1])
A2B = np.matmul(np.matmul(A,A), B)
AB = np.matmul(A, B)

P = np.hstack((B, AB, A2B))
np.linalg.matrix_rank(P)

M = np.array([[1,0,1],
              [1, 1,0],
              [1,2,0]])
np.linalg.matrix_rank(M)

# np.linalg.inv(M)
A_hat = np.matmul(np.matmul(np.linalg.inv(M), A), M)
B_hat = np.matmul(np.linalg.inv(M), B)
C_hat = np.matmul(C, M)

M_prime = np.linalg.inv(np.array([[3,3], [1,0]]))
```

```python
# q5.a
A = np.array([[0,1],
              [0,0]])
np.linalg.eig(A)

# q5.c
x1 = np.linspace(-10, 10, 200)
x2 = np.linspace(-10, 10, 200)
X1, X2 = np.meshgrid(x1, x2)

X1dot = X2 - X1*(X2**2)
X2dot = -(X1**3)

X1_lin, X2_lin = np.meshgrid(np.linspace(-1, 1, 20), np.linspace(-1, 1, 20))
X1dot_lin = X2_lin
X2dot_lin = 0*X2_lin



fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

ax1.streamplot(X1, X2, X1dot, X2dot)
ax1.set_title('Non-Linear System Phase Portrait')
ax1.set_xlabel('$x_1$')
ax1.set_ylabel('$x_2$')


ax2.streamplot(X1_lin, X2_lin, X1dot_lin, X2dot_lin)
ax2.set_title('Linear System Phase Portrait')
ax2.set_xlabel('$x_1$')
ax2.set_ylabel('$x_2$')

# q5.d
Vdot = -4*(X1**4)*(X2**2)
# Create a figure and 3D axes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X1, X2, Vdot, cmap='viridis', edgecolor='none')

# Set labels and title
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$Vdot(x_1, x_2)$')
ax.set_title('Vdot v/s X1 and X2')
```
